```python
import re
import string
from collections import Counter
```

1. Difference between "Love" and "love":---->
In NLP, "Love" and "love" are treated as different tokens if case normalization is not applied.
This increases vocabulary size and may reduce model efficiency. Lowercasing ensures consistency.

2. What happens if stopwords are not removed?
Stopwords add noise and increase computational cost. Models may focus on less meaningful words.

3. When removing stopwords is harmful:
- Sentiment Analysis: "not good" → removing "not" changes meaning
- Question Answering: "What is AI?" → removing "what" loses intent

4. Stemming vs Lemmatization:
- Stemming: Cuts words crudely (running → run)
- Lemmatization: Uses vocabulary + grammar (better accuracy)

```python
#Preprocessing Function (Task 2)

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""
    # Remove URLs and emails
    text = re.sub(r"http\S+|www\S+|https\S+", "", text)
    text = re.sub(r"\S+@\S+", "", text)
    # Remove numbers
    text = re.sub(r"\d+", "", text)

    # Convert to lowercase
    text = text.lower()
    # Remove repeated characters (soooo → so)
    text = re.sub(r'(.)\1{2,}', r'\1', text)
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()
    # Tokenization
    tokens = text.split()
    # Remove short tokens (≤2 except "no", "not")
    tokens = [word for word in tokens if len(word) > 2 or word in ["no", "not"]]
    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

```python
# Stress Testing (Task 3)
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    results.append((sentence, tokens, clean))

    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-" * 50)
```

```python
#Token Analytics (Task 4)

analytics = []

for sentence, tokens, clean in results:
    if len(tokens) == 0:
        avg_len = 0
    else:
        avg_len = sum(len(t) for t in tokens) / len(tokens)

    analytics.append({
        "sentence": sentence,
        "total_tokens": len(tokens),
        "unique_tokens": len(set(tokens)),
        "avg_token_length": round(avg_len, 2)
    })

for a in analytics:
    print(a)
```

Most noisy sentence:--->
"Win $$$ now!!! Limited offer!!!" → symbols + repetition

Most meaningful after cleaning:--->
"I am not happy with this" → retains sentiment meaning clearly

```python
#Frequency Analysis (Task 5)

all_tokens = []

for _, tokens, _ in results:
    all_tokens.extend(tokens)

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
#Full Pipeline (Task 6)

def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }

pipeline_output = full_pipeline(test_sentences)
print(pipeline_output)
```

```python
#Error Handling (Task 7)

edge_cases = [
    "",
    "😂😂😂😂",
    "1234567890"
]

for case in edge_cases:
    tokens, clean = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-" * 40)
```
